## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns = ["EIN", "NAME"])
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
counts = application_df["APPLICATION_TYPE"].value_counts()
counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = counts[counts < 700].index.tolist()
#Increase Our cutoff values to further reduce our dementions

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_counts = application_df["CLASSIFICATION"].value_counts()
classification_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
counts_greater = classification_counts[classification_counts > 1]
counts_greater
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = classification_counts[classification_counts < 2000].index.tolist()
#Increase Our cutoff values to further reduce our dementions

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
test_df = pd.get_dummies(application_df)
test_df
```

```python
# Split our preprocessed data into our features and target arrays
y = test_df["IS_SUCCESSFUL"]
x = test_df.drop(columns = ["IS_SUCCESSFUL", "SPECIAL_CONSIDERATIONS_N", "SPECIAL_CONSIDERATIONS_Y"])

#Dropping columns to reduce our target array
                 
print(len(x.columns))

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state = 42)

#Mess with our random_state
```

```python
# Create a StandardScaler instance
scaler = StandardScaler()

# Fit and transform the StandardScaler on training data
X_train_scaled = scaler.fit_transform(X_train)

# Transform the testing data using the same scaler
X_test_scaled = scaler.transform(X_test)

# Now, let's do Min-Max Scaling on the scaled data
min_max_scaler = MinMaxScaler()

# Fit and transform the MinMaxScaler on scaled training data
X_train_scaled_normalized = min_max_scaler.fit_transform(X_train_scaled)

# Transform the scaled testing data using the same MinMaxScaler
X_test_scaled_normalized = min_max_scaler.transform(X_test_scaled)

#Scaled and Normalized data
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units = 160, activation = "relu", input_dim = 38))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units = 120, activation = "relu"))

nn.add(tf.keras.layers.Dense(units = 80, activation = "relu"))

nn.add(tf.keras.layers.Dense(units = 40, activation = "relu"))

nn.add(tf.keras.layers.Dense(units = 20, activation = "relu"))

nn.add(tf.keras.layers.Dense(units = 10, activation = "relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units = 1, activation = "sigmoid"))

#Increasing the number of layers and neurons

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
model = nn.fit(X_train_scaled_normalized, y_train, epochs=700)

#Run for an absurd amount of epochs
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled_normalized,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
